````python
// === Проверка: резолвится ли домен из JVM ===
import java.net.InetAddress

def checkHost(hostname: String): Unit = {
  try {
    val addr = InetAddress.getByName(hostname)
    println(s"✅ $hostname → ${addr.getHostAddress}")
  } catch {
    case e: Exception => 
      println(s"❌ Cannot resolve $hostname: ${e.getMessage}")
  }
}

// Проверьте оба адреса
checkHost("nessie.local")                    // ✅ Должно работать
checkHost("nessie.nessie.svc.cluster.local") // ❌ Скорее всего, не сработает с хоста
````
